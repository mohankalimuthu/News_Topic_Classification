```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import joblib
```

```python
train_df = pd.read_csv("C:\\Users\\hp\\PyCharmMiscProject\\machine learning\\Algorithms\\Naive bayes\\MNB\\News_topic_Classification\\Data\\training_data.csv")
```

```python
print(train_df)
```

```python
train_df['label'].unique()
```

```python
test_df = pd.read_csv("C:\\Users\\hp\\PyCharmMiscProject\\machine learning\\Algorithms\\Naive bayes\\MNB\\News_topic_Classification\\Data\\test_data.csv",names=["text","label"])
```

```python
test_df
```

```python
label_mapping = {0:'World',1:'Sports',2:'Business',3:'Sci/Tech'}
train_df['category'] = train_df['label'].map(label_mapping)
test_df['category'] = test_df['label'].map(label_mapping)
```

```python
train_df.info()
```

```python
x = train_df['text']
y = train_df['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42,stratify=y)
```

```python
model = Pipeline([
    ("tfidf",TfidfVectorizer(stop_words = 'english')),
    ("nb",MultinomialNB())
])
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
